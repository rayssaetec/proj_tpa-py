## Atividade 11 - Estudo de Caso Supermercado

```python
class Supermarket:
    def __init__(self):
        #Atributo 'fila' é uma lista vazia
        self.caixarapido = []
        self.caixaidosos = []
        self.caixanormal = []

    def adiciona(self, nome, idade, unidade):
        #Adiciona o 'a pessoa' como atributo do respectivo 'caixa'

        if idade < 60 and unidade <= 20:
            self.caixarapido.append(nome)
            print(f'{nome} chegou no caixa rápido')

        elif idade >= 60:
            self.caixaidosos.append(nome)
            print(f'{nome} chegou no caixa de idosos')

        else:
            self.caixanormal.append(nome)
            print(f'{nome} chegou no caixa comum')

    def vazio_normal(self):
        #Retorno True se o comprimento da lista for vazio e False caso acontrário
        return len(self.caixanormal) == 0

    def vazio_rápido(self):
        #Retorno True se o comprimento da lista for vazio e False caso acontrário
        return len(self.caixarapido) == 0
    
    def vazio_idosos(self):
        #Retorno True se o comprimento da lista for vazio e False caso acontrário
        return len(self.caixaidosos) == 0
    
    def atendido_caixa_normal(self):
        if not self.vazio_normal():
            #Remove com o método pop, o primeiro nome da lista 'fila'
            nome = self.caixanormal.pop(0)
            print(f'{nome} foi atendido no caixa comum')
            return nome
        else:
            print('A fila está vazia')
            return None
        
    def atendido_caixa_rapido(self):
        if not self.vazio_rápido():
            #Remove com o método pop, o primeiro nome da lista 'fila'
            nome = self.caixarapido.pop(0)
            print(f'{nome} foi atendido no caixa rápido')
            return nome
        else:
            print('A fila está vazia')
            return None

    def atendido__caixa_idosos(self):
        if not self.vazio_idosos():
            #Remove com o método pop, o primeiro nome da lista 'fila'
            nome = self.caixaidosos.pop(0)
            print(f'{nome} foi atendido no caixa de idosos')
            return None
        else:
            print('A fila está vazia')
            return None
    
    def tamanho_caixa_normal(self):
        if len(self.caixanormal)==0:
            print('O caixa está livre')
        else:
            print(f'Há somente {len(self.caixanormal)} pessoas na fila do caixa comum')
    
    def tamanho_caixa_rapido(self):
        if len(self.caixarapido)==0:
            print('O caixa está livre')
        else:
            print(f'Há somente {len(self.caixarapido)} pessoas na fila do caixa rápido')

    def tamanho_caixa_idosos(self):
        if len(self.caixaidosos)==0:
            print('O caixa está livre')
        else:
            print(f'Há somente {len(self.caixaidosos)} pessoas na fila do caixa de idosos')
```

```python
fila = Supermarket()
fila.adiciona('Maria', 34, 29)
fila.adiciona('Cristian', 60, 10)
fila.adiciona('Rayssa', 18, 20)
fila.adiciona('Regina', 13, 50)
fila.adiciona('Wilson', 85, 10)
fila.adiciona('Marcelo', 18, 23)
fila.adiciona('Paulo', 62, 29)
fila.adiciona('Gabriel', 25, 10)
fila.adiciona('Fernanda', 18, 200)
fila.adiciona('Micheli', 23, 20)
fila.atendido_caixa_normal()
fila.atendido_caixa_rapido()
fila.atendido__caixa_idosos()
fila.tamanho_caixa_idosos()
fila.tamanho_caixa_normal()
fila.tamanho_caixa_rapido()
```
